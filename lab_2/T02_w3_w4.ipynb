````python
def find_intersection(a, b, c, d):
    if b < c or a > d:
        return "–û—Ç—Ä–µ–∑–∫–∏ –Ω–µ –ø–µ—Ä–µ—Å–µ–∫–∞—é—Ç—Å—è"
    else:
        start = max(a, c)
        end = min(b, d)
        return [start,end]

intersection = find_intersection(6, 8, 3, 2)
if isinstance(intersection, str):
    print(intersection)
else:
    print("–ü–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ:", intersection)
````

# 8
–ù–∞–ø–∏—à–∏—Ç–µ –ø—Ä–æ–≥—Ä–∞–º–º—É –¥–ª—è —Ä–µ—à–µ–Ω–∏—è —É—Ä–∞–≤–Ω–µ–Ω–∏—è
$$ ax^2+bx+c=0 $$
–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã $a, b, c$ –º–æ–≥—É—Ç –±—ã—Ç—å –ª—é–±—ã–º–∏ —á–∏—Å–ª–∞–º–∏.

````python
import math
 
a = float(input("a = "))
b = float(input("b = "))
c = float(input("c = "))
 
discr = b ** 2 - 4 * a * c
print("–î–∏—Å–∫—Ä–∏–º–∏–Ω–∞–Ω—Ç D = %.2f" % discr)
 
if discr > 0:
    x1 = (-b + math.sqrt(discr)) / (2 * a)
    x2 = (-b - math.sqrt(discr)) / (2 * a)
    print("x1 = %.2f \nx2 = %.2f" % (x1, x2))
elif discr == 0:
    x = -b / (2 * a)
    print("x = %.2f" % x)
else:
    print("–ö–æ—Ä–Ω–µ–π –Ω–µ—Ç")
````

# 19
–î–∞–Ω—ã m (m>1) —Å–ª–æ–≤. –í –∫–∞–∫–æ–º –∏–∑ –Ω–∏—Ö –¥–æ–ª—è (–≤ %) –∑–∞–¥–∞–Ω–Ω–æ–π –±—É–∫–≤—ã –±–æ–ª—å—à–µ.

````python
def calculate_percentage(word, letter):
    count = 0
    for ch in word:
        if ch == letter:
            count += 1
    percentage = (count / len(word)) * 100
    return percentage


m = int(input("–í–≤–µ–¥–∏—Ç–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Å–ª–æ–≤: "))
letter = input("–í–≤–µ–¥–∏—Ç–µ –∑–∞–¥–∞–Ω–Ω—É—é –±—É–∫–≤—É: ")

max_percentage = 0
max_word = ""

for i in range(m):
    word = input("–í–≤–µ–¥–∏—Ç–µ —Å–ª–æ–≤–æ: ")
    percentage = calculate_percentage(word, letter)
    if percentage > max_percentage:
        max_percentage = percentage
        max_word = word

print("–°–ª–æ–≤–æ —Å –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–π –¥–æ–ª–µ–π –∑–∞–¥–∞–Ω–Ω–æ–π –±—É–∫–≤—ã:", max_word)
print("–î–æ–ª—è –∑–∞–¥–∞–Ω–Ω–æ–π –±—É–∫–≤—ã –≤ —ç—Ç–æ–º —Å–ª–æ–≤–µ:", max_percentage, "%")
````
